# Bronx Subway Network Graph

This notebook creates a simple network graph using Python's `NetworkX` package.  
For this example, I modeled a section of the **MTA 4 train in the Bronx**:

- **Nodes** represent subway stations (e.g., Woodlawn, Fordham Rd, 161st-Yankee Stadium).  
- **Edges** represent direct connections between stations (adjacent stops along the line).  

The goal is to demonstrate how to build and visualize a basic graph by:  
1. Adding nodes (stations).  
2. Adding edges (connections between stations).  
3. Drawing the network using `matplotlib`.  

This serves as a "beachhead program" for getting familiar with graphs in Python and can be extended in later projects.


```python
#Attempting to create my own graph 
bronx_subway = nx.Graph()

stations = [
    "Woodlawn", "Mosholu Parkway", "Bedford Park Blvd-Lehman College", 
    "Kingsbridge Rd", "Fordham Rd", "183rd St", "Burnside Ave", 
    "176th St", "Mt Eden Ave", "170th St", "167th St", "161st St-Yankee Stadium"
]

bronx_subway.add_nodes_from(stations)

for i in range(len(stations)-1):
    bronx_subway.add_edge(stations[i], stations[i+1])

plt.figure(figsize=(10,6))
nx.draw(
    bronx_subway,
    with_labels=True, 
    node_color="pink",
)

plt.title("Bronx MTA 4 Train")
plt.show()
```
